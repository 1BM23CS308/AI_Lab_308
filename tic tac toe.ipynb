```python
# Tic Tac Toe game in Python (2-player mode)

# Function to display the board
def print_board(board):
    print("\n")
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print("\n")

# Function to check for a winner
def check_winner(board, player):
    win_positions = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
        (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
        (0, 4, 8), (2, 4, 6)              # Diagonals
    ]
    for a, b, c in win_positions:
        if board[a] == board[b] == board[c] == player:
            return True
    return False

# Main function to play the game
def tic_tac_toe():
    print("üéÆ Welcome to Tic Tac Toe!")
    print("Player 1 = X | Player 2 = O")
    print("Enter positions as numbers 1‚Äì9 like this:")
    print("1 | 2 | 3\n4 | 5 | 6\n7 | 8 | 9\n")

    board = [' '] * 9
    current_player = 'X'
    moves = 0

    # Game loop
    while True:
        print_board(board)
        try:
            choice = int(input(f"Player {current_player}, choose your position (1-9): ")) - 1
            if choice < 0 or choice > 8:
                print("‚ö†Ô∏è Invalid input! Choose a number between 1 and 9.")
                continue
            if board[choice] != ' ':
                print("‚ö†Ô∏è That spot is already taken! Try again.")
                continue
        except ValueError:
            print("‚ö†Ô∏è Please enter a valid number.")
            continue

        # Place the move
        board[choice] = current_player
        moves += 1

        # Check for win
        if check_winner(board, current_player):
            print_board(board)
            print(f"üèÜ Player {current_player} wins the game!")
            break

        # Check for draw
        if moves == 9:
            print_board(board)
            print("ü§ù It's a draw!")
            break

        # Switch players
        current_player = 'O' if current_player == 'X' else 'X'

# Run the game
if __name__ == "__main__":
    tic_tac_toe()
```
